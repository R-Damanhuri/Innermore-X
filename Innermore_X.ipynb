```python
import re

def clean(text):
    text = text.lower()
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'@\w+\s?', '', text)
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
    text = ' '.join(text.split())

    return text
```

```python
def merge_tokens(df_sentences, df_tokens):
    token_lists = []
    tag_lists = []
    idx_start = 0

    for i in range(len(df_sentences)):
        length = len(str(df_sentences['full_text'][i]).split())
        idx_end = idx_start + length
        token_list = [token for token in df_tokens['full_text'][idx_start:idx_end]]
        tag_list = [token for token in df_tokens['label'][idx_start:idx_end]]

        token_lists.append(token_list)
        tag_lists.append(tag_list)

        idx_start = idx_end

    new_df = pd.DataFrame({'tokens': token_lists, 'tags': tag_lists})
    return new_df
```

```python
from transformers import AutoTokenizer
from transformers import DataCollatorForTokenClassification
from transformers import TFAutoModelForTokenClassification
import evaluate
import numpy as np
from transformers.keras_callbacks import KerasMetricCallback
from transformers import create_optimizer
import tensorflow as tf
from transformers.keras_callbacks import PushToHubCallback
```

```python
def align_labels_with_tokens(labels, word_ids):
    new_labels = []
    current_word = None
    for word_id in word_ids:
        if word_id != current_word:
            # Start of a new word!
            current_word = word_id
            label = -100 if word_id is None else labels[word_id]
            new_labels.append(label)
        elif word_id is None:
            # Special token
            new_labels.append(-100)
        else:
            # Same word as previous token
            label = labels[word_id]
            # If the label is B-XXX we change it to I-XXX
            if label % 2 == 1:
                label += 1
            new_labels.append(label)

    return new_labels
```

```python
def tokenize_and_align_labels(examples):
    tokenized_inputs = tokenizer(examples["tokens"], is_split_into_words=True)
    all_labels = examples["tags"]
    new_labels = []
    for i, labels in enumerate(all_labels):
        word_ids = tokenized_inputs.word_ids(i)
        new_labels.append(align_labels_with_tokens(labels, word_ids))

    tokenized_inputs["labels"] = new_labels
    return tokenized_inputs
```

```python
metric = evaluate.load("seqeval")
```

```python
def compute_metrics(p):
    predictions, labels = p
    predictions = np.argmax(predictions, axis=2)

    true_predictions = [
        [label_names[p] for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]
    true_labels = [
        [label_names[l] for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]

    results = metric.compute(predictions=true_predictions, references=true_labels)
    return {
        "val_precision": results["overall_precision"],
        "val_recall": results["overall_recall"],
        "val_f1": results["overall_f1"],
        "val_accuracy": results["overall_accuracy"],
    }
```

```python
id2label = {i: label for i, label in enumerate(label_names)}
label2id = {v: k for k, v in id2label.items()}
```

## indobert-base-p1

```python
model_name = "indobert-base-p1"
model_path = "indobenchmark/indobert-base-p1"
```

```python
print(f"Finetuning {model_name}: Start!")
tokenizer = AutoTokenizer.from_pretrained(model_path)
innermore_tokenized = innermore_dict.map(
  tokenize_and_align_labels,
  batched=True,
  remove_columns=innermore_dict["train"].column_names,
)

data_collator = DataCollatorForTokenClassification(
  tokenizer=tokenizer, return_tensors="tf"
)

tf_train_dataset = innermore_tokenized["train"].to_tf_dataset(
  columns=["attention_mask", "input_ids", "labels", "token_type_ids"],
  collate_fn=data_collator,
  shuffle=True,
  batch_size=16,
)

tf_eval_dataset = innermore_tokenized["validation"].to_tf_dataset(
  columns=["attention_mask", "input_ids", "labels", "token_type_ids"],
  collate_fn=data_collator,
  shuffle=False,
  batch_size=16,
)

model = TFAutoModelForTokenClassification.from_pretrained(
  model_path,
  id2label=id2label,
  label2id=label2id,
  from_pt=False
)

tf.keras.mixed_precision.set_global_policy("mixed_float16")
num_epochs = 15
num_train_steps = len(tf_train_dataset) * num_epochs

optimizer, schedule = create_optimizer(
  init_lr=2e-4,
  num_warmup_steps=0,
  num_train_steps=num_train_steps,
  weight_decay_rate=0.01,
)
model.compile(optimizer=optimizer)

metric_callback = KerasMetricCallback(metric_fn=compute_metrics, eval_dataset=tf_eval_dataset)
push_to_hub_callback = PushToHubCallback(output_dir=f"innermore-x-{model_name}", tokenizer=tokenizer, save_strategy="no")

history = model.fit(
  tf_train_dataset,
  validation_data=tf_eval_dataset,
  callbacks=[metric_callback, push_to_hub_callback],
  epochs=num_epochs,
)
np.save('history_'+ model_name +'.npy', history.history)
print(f"Finetuning {model_name}: Done!\n")
```

```python
def decode_label(input_token, token_label):
    word_labels = []
    for i, token in enumerate(input_token):
        if token.startswith("["):
            continue
        if token.startswith("##"):
            continue
        word_labels.append(token_label[i])
    return word_labels
```

```python
def decode_labels_batch(pred_token_labels, input_tokens_batch):
    batch_word_labels = []
    for pred_token_label, input_tokens in zip(pred_token_labels, input_tokens_batch):
        word_labels = decode_label(input_tokens, pred_token_label)
        batch_word_labels.append(word_labels)
    return batch_word_labels
```

```python
# inputs[0].tokens = token sentence ke-0, termasuk special token
# predicted_token_class[0] = label token sentence ke-0, termasuk special token
# tokenizer.decode(inputs['input_ids'][0]) = kata-kata sentence ke-0
```

```python
input_token = inputs[0].tokens
token_label = pred_token_labels[0]

print(input_token)
print(token_label)
```

```python
def decode_label(input_token, token_label):
    word_labels = []
    for i, token in enumerate(input_token):
        if token.startswith("["):
            continue
        if token.startswith("##"):
            continue
        word_labels.append(token_label[i])
    return word_labels
```

```python
def decode_labels_batch(pred_token_labels, input_tokens_batch):
    batch_word_labels = []
    for pred_token_label, input_tokens in zip(pred_token_labels, input_tokens_batch):
        word_labels = decode_label(input_tokens, pred_token_label)
        batch_word_labels.append(word_labels)
    return batch_word_labels
```
